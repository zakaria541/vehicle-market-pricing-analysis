```python
## Plot a histogram of distrbution of vehicle prices, label x-axis as Price (USD) and y-axis as frequency, figure size (10,6),bins=50.
import matplotlib.pyplot as plt
plt.figure(figsize=(10, 6))
plt.title('Distribution of Vehicle Prices')
plt.hist(df['price'], bins=50, color= 'blue', edgecolor='black')
plt.xlabel('Price (USD)')
plt.ylabel('Frequency')
plt.show()
```

```python
## Create a countplot for distribution of manufacturers, figure size (12,6)
import seaborn as sns
plt.figure(figsize=(12, 6))
sns.countplot(x='make', data=df, color='blue', edgecolor='black')
plt.xlabel('Manufacturer')
plt.ylabel('Count')
plt.xticks(rotation=90)
plt.title('Distribution of Manufacturers') 
plt.show()
```

```python
## Create a bar chart of price against number of doors
cleaned_data = df[['price', 'doors']].dropna() 
price_by_doors = cleaned_data.groupby('doors')['price'].mean()
plt.figure(figsize=(10,6)) 
price_by_doors.plot(kind='bar', color='blue', edgecolor='black') 
plt.title('Average Price by Number of Doors')
plt.xlabel('Number of Doors') 
plt.ylabel('Average Price (in USD)') 
plt.xticks(rotation=45) 
plt.show() 
```

# Comment on the chart of price against number of doors
 This graph gives an overview of average prices in USD concerning the number of doors of cars. It can be seen from this view that the average price of five-door cars is nearly 80,000 dollars. Two and three-door automobiles have similar average prices, which have been within a bracket of 55,000 dollars to 60,000 dollars, while four-door automobiles have an average, just below  50,000 dollars.

```python
## Create a bar chart of price against cylinders
cleaned_data_cylinders = df[['price', 'cylinders']].dropna()
price_by_cylinders = cleaned_data_cylinders.groupby('cylinders')['price'].mean()
plt.figure(figsize=(10,6))
price_by_cylinders.plot(kind='bar', color='yellow', edgecolor='blue')
plt.title('Average Price by Number of Cylinders')
plt.xlabel('Number of Cylinders') 
plt.ylabel('Average Price (in USD)') 
plt.xticks(rotation=0)
plt.show()
```

# Comment on the bar chart of price against cylinders
Below is a histogram showing the mean price in USD of cars against the number of cylinders it possesses. The most expensive cars are those with zero cylinders at about 65,000 dollars. These are followed by six and eight-cylinder cars at just over 60,000 dollars at 55,000 dollars respectively. The three-cylinder cars are the cheapest, averaging just over 30,000; four falls in the middle at just less than 50,000. It may suggest that the smaller, more fuel-efficient cars would be cheaper while the ore expensive would then be the ones with more cylinders or combustion engines without a restrictor.


```python
## Create a bar chart of price against mileage
plt.figure(figsize=(18,12)) 
price_by_mileage=df.groupby('mileage')['price'].mean() 
price_by_mileage.plot(kind='bar', color='blue', edgecolor='black') 
plt.xlabel('mileage')
plt.ylabel('Price')
plt.title('Price vs Mileage')
plt.xticks(rotation=90)
plt.show()
```

# Comment on the bar chart of price against mileage
 This bar chart shows the relationship between price and mileage. The tall bars on the left, where the mileage is smaller and the prices rocket at 175,000, show that higher prices are normally associated with low mileage. Although there are still occasional spikes of higher pricing, for the most part, as mileage increases-which it does mostly between 25,000 and 100,000-the prices bunch up around the lower values. Although there are a number of notable exceptions scattered through the data, the overall trend of the figure is inverse, with costs decreasing as mileage increases, which suggests that other factors may also have an impact on price.



```python
## Create a box plot of price
plt.figure(figsize=(20,10))
sns.boxplot(data=df, y='price')
plt.title('Box plot of price')
plt.show()
```

# Comment on the box plot of price
This box plot represents the distribution of prices by showing the central tendency as well as the presence of outliers. The median price is depicted in the box with an accumulation of about 50,000, while the box itself shows the interquartile range. The whiskers show the range for the majority of the data, from about 25,000 on the lower extreme to over 100,000 on the high extreme. There are a number of outliers above 100,000, some of them as high as 200,000, indicating there might be a number of really high values. There is also one low outlier close to zero. Overall, the data shows consistency of range, with a few notable wobbles in price.

```python
##Create a countplot of exterior color with figure size(6,45)
plt.figure(figsize=(6, 45)) 
sns.countplot(y='exterior_color', data=df, color='blue', edgecolor='black') 
plt.title('Count of vehicles by exterior color')
plt.show()
```

```python
# Create a bar chart of price against year
price_by_year=df.groupby('year')['price'].mean()
plt.figure(figsize=(10, 6))
price_by_year.plot(kind='bar', color='blue', edgecolor='black')
plt.xlabel('Year')
plt.ylabel('averagePrice')
plt.title('Average Price by Year')
plt.xticks(rotation=0)
plt.show()
```

# Comment on the bar chart of price against year
In the "Average Price by Year" bar chart, the values from 2023 to 2025 remain quite constant at about 50,000. Although from this steep increase in 2025, no motion is really noticeable, showing that the price constantly stays. The blue bars are good to visualize such stability.
